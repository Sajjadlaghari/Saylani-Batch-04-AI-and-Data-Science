# Arthematics in pandas

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2['sum'] = var2['A']+var["B"]

var2
```

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2['Sub'] = var2['A']-var["B"]

var2
```

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2['div'] = var2['A']/var["B"]

var2
```

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2['multiplication'] = var2['A']*var["B"]

var2
```

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2['Remender'] = var2['A']%var["B"]

var2
```

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2['Greader From A'] = var2['A'] < 4

var2
```

# Insert New Column in DataFrame

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2.insert(1,'Greader', var2['A'])

var2
```

```python

var2 = pd.DataFrame(    {'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2.insert(1,'Greader', [5,10,15,20])

var2
```

```python

var2 = pd.DataFrame({'A':[1,2,3,4],'C':[5,6,7,8],'B':[5,6,7,8]})

var2.insert(1,'X' ,var2['A'])
var2.insert(2,'Y' ,var2['A'])
var2.insert(1,'vv' ,var2['B'])

var2
```
