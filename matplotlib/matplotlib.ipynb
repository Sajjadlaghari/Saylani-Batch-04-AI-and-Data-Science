# Bar Plot

```python
import matplotlib.pyplot as plt

x=['java','python','c','ruby']
y=[85,95,40,83]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('XYZ')
plt.title('Lanuages Ranking')
c=['b','g','y','purple']
plt.bar(x,y,width=0.4,color=c,edgecolor='orange',linewidth=1)

plt.show()

```

# Pie Plot

```python
x=[20,60,10,15]
y=['Java','Python','Ruby Rails',"React Js"]
ex=[0.1,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True)
```
