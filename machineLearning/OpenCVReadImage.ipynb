```python
import cv2


img = cv2.imread('image.jpeg')

cv2.imshow('My Image',img)
key = cv2.waitKey(0)

if key == 27:
    cv2.destroyAllWindows()
elif key == ord('s'):
    cv2.imwrite('my_copy_image.png', img)
    cv2.destroyAllWindows()
    
```
